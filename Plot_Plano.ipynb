```python
# --- 1. Upload Excel file ---
from google.colab import files
uploaded = files.upload()

import io
import pandas as pd
import numpy as np

# Read the first uploaded file
filename = list(uploaded.keys())[0]
df = pd.read_excel(io.BytesIO(uploaded[filename]), header=None)

# Convert to numpy array
Z = df.values

# --- 2. Build coordinate grid ---
rows, cols = Z.shape
X, Y = np.meshgrid(np.arange(cols), np.arange(rows))

# --- 3. Plot the surface ---
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap='viridis', rstride=1, cstride=1, linewidth=0)

ax.set_title("Surface Plot from Excel Matrix", fontsize=14)
ax.set_xlabel("X index")
ax.set_ylabel("Y index")
ax.set_zlabel("Value")

plt.show()
```
